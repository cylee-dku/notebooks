## Import libraries

```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd

# This is from the project library

from wafermaps import plot_wafermaps
```

## Read dataset and image folder

```python
# Read the dataset as a Pandas dataframe in memory
# Note: here, we can choose to read the first 100K rows. Other sampling options are available
dataset_wafermaps_prepared = dataiku.Dataset("wafermaps_prepared")
df = dataset_wafermaps_prepared.get_dataframe(limit=100000)

# Image folder
images = dataiku.Folder('kGsrXVzy')
file_paths = images.list_paths_in_partition()
print("Number of files: " + str(len(file_paths)) + ". Printing first 10.")
file_paths[:10]
```

```python
# Get some simple descriptive statistics
pdu.audit(df)
```

## List the defect types

```python
defect_types = df.defect_type.unique()
print(defect_types)
```

## Plot wafermaps

```python
plot_wafermaps(df, num_columns=5, defect_types=defect_types, img_folder=images)
```
