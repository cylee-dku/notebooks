```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd
```

```python
# Read the dataset as a Pandas dataframe in memory
# Note: here, we only read the first 100K rows. Other sampling options are available
dataset_quality_analysis_tall = dataiku.Dataset("quality_analysis_tall")
df = dataset_quality_analysis_tall.get_dataframe()
```

```python
# Get some simple descriptive statistics
pdu.audit(df)
```

```python
# Concatenate sensor and timestamp for pivoting
df['sensor_timestamp'] = df.sensor.astype(str) + "_" + df.timestamp.astype(str)
df_sorted = df.sort_values(by = ['sensor', 'timestamp'])
```

```python
# Pivot
pivoted = df_sorted.pivot(index='run_wafer', columns='sensor_timestamp', values='value')
pivoted.head(10)
```
